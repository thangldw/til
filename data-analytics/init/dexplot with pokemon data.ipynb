```python
import dexplot as dxp
import pandas as pd

df = pd.read_csv("datasets_121_280_Pokemon.csv")
```

```python
df.head()
```

```python
dxp.bar(x='Type 1', y='Total', data=df, aggfunc='median', sort_values='asc')
```

```python
dxp.bar(x='Type 1', y='Total', data=df, aggfunc='median', x_order='asc', split='Generation')
```

```python
dxp.line(x='Type 1', y='HP', data=df, aggfunc='median', sort_values='desc')
```

```python
dxp.scatter(x='Type 1', y='Attack', data=df, aggfunc='median')
```

```python
df.groupby('Type 1').agg({'Speed': 'median'})
```

```python
df['Type 2'].value_counts()
```

```python
df.pivot_table(index='Type 1', columns='Type 2', values='Attack', aggfunc='median')
```

```python
dxp.bar(x='Type 1', y='Total', data=df, aggfunc='median', x_order='asc', split='Legendary', stacked=True)
```

```python
dxp.bar(x='Type 1', y='Total', data=df, aggfunc='median', x_order='asc', split='Legendary', 
        col='Type 2', col_order=['Electric', 'Dragon', 'Grass'], size=5)
```

```python
dxp.box(x='Attack', y='Type 1', data=df, split='Legendary',  x_order=[50, 100])
```

```python
dxp.hist(val='Type 1', data=df)
```

```python
dxp.scatter(x='Attack', y='Defense', data=df, 
            split='Type 1', col='Generation', col_order=[2, 3])
```
